```python
import numpy as np
import matplotlib.pyplot as plt
#to activate matplotlib
%matplotlib inline 
#image function, allow us to transform image that numpy can understand
from PIL import Image
```

```python
pic = Image.open('./DATA/00-puppy.jpg')
plt.imshow(pic) #this is used to display the image in jupyter notebook
plt.show()
type(pic)
#numpy can not understand PIL.JpegImagePlugin.JpegImageFile
```

```python
pic_array = np.asarray(pic)     #this will convert the pic into np array which could understand by numpy
type(pic_array)
```

```python
plt.imshow(pic_array) 
```

```python
pic_array.shape #the 3 at the end mean 3 color chanels, RGB colors
```

```python
picRGB = pic_array.copy()
#in the 3 color chanels: R G B, R is at index 0...
pic_red = picRGB[:,:,0] #thie will get all Red color contribution
plt.imshow(pic_red, cmap="gray")#show im gray scale, color mapping, closer to gray, more red
```

```python
pic_green = picRGB[:,:,1] #thie will get all Green color contribution
plt.imshow(pic_green,cmap="gray")
```

```python
pic_blue = picRGB[:,:,2] #thie will get all Blue color contribution
plt.imshow(pic_blue,cmap="gray")
```

```python
pic_removeGreen = picRGB.copy()
pic_removeGreen[:,:,1] = 0  #this will remove all green color
plt.imshow(pic_removeGreen)
```

```python
pic_removeBlue = picRGB.copy()
pic_removeBlue[:,:,2] = 0  #this will remove all blue color
plt.imshow(pic_removeBlue)
```

```python
#remove R G B
pic_removeBlue[:,:,0] = 0
pic_removeBlue[:,:,1] = 0
plt.imshow(pic_removeBlue)
```
