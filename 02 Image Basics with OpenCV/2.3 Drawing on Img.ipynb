```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
print(blank_img.shape)
```

```python
plt.imshow(blank_img)
#this is black because created using np,zeros()
```

```python
#add a rectangular to the above img
cv2.rectangle(blank_img,pt1=(384,0), pt2=(510,150),color=(0,255,0),thickness=10)
#thickness 参数定义了矩形边界的宽度。这里 thickness=10 表示边界宽度为10个像素。
#pt1 is the top left cornor
#pt2 is the right bottom cornor
```

```python
plt.imshow(blank_img)
#thickness 参数定义了矩形边界的宽度。这里 thickness=10 表示边界宽度为10个像素。
#因为上面和右侧已经到达边界，所以显示出来会窄
```

```python
cv2.rectangle(blank_img,pt1=(200,200), pt2=(300,300),color=(0,0,255),thickness=10)
plt.imshow(blank_img)
```

```python
#add circle
cv2.circle(img=blank_img,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(blank_img)
```

```python
#add a filled circle
cv2.circle(img=blank_img,center=(400,400),radius=50,color=(255,0,0),thickness=-1)
plt.imshow(blank_img)
```

```python
#draw line
cv2.line(blank_img,pt1=(0,0),pt2=(512,512),color=(102,255,255),thickness=5)
plt.imshow(blank_img)
```

```python
#add text
font = cv2.FONT_HERSHEY_SIMPLEX
#org=(start at x = 10, start at y = 500)
cv2.putText(blank_img,text="Hello!",org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
blank_img2 = np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(blank_img2)
```

```python
#draw polygons
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
vertices.shape
#the shape is (4,2) which is 2D, but OpenCV want a 3D, we are missing the color channel
#reshape is to 3D
pts = vertices.reshape((-1,1,2))
print(vertices)
print(vertices.shape)
print("----------------------------------------")
print(pts)
print(pts.shape)
```

```python
#draw polygons
cv2.polylines(blank_img2,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(blank_img2)
```
