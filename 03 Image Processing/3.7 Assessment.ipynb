```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
imgBGR = cv2.imread("../DATA/giraffes.jpg")
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
display(imgRGB)
```

```python
#apply threshold
ret, thresh1 = cv2.threshold(imgRGB,thresh=100,maxval=255,type=cv2.THRESH_BINARY)
display(thresh1)
```

```python
#if I want the threshold img in gray scale
imgBGR = cv2.imread("../DATA/giraffes.jpg",0)
#flags 的值等于0表示以灰度模式读取图像
ret, thresh2 = cv2.threshold(imgBGR,thresh=100,maxval=255,type=cv2.THRESH_BINARY)
display(thresh2,cmap="gray")
```

```python
imgBGR = cv2.imread("../DATA/giraffes.jpg")
imgHSV = cv2.cvtColor(imgBGR,cv2.COLOR_BGR2HSV)
display(imgHSV)
```

```python
#build kernel for blur the img
kernel = np.ones((4,4),dtype=np.float32) / 10
print(kernel)
```

```python
imgBGR = cv2.imread("../DATA/giraffes.jpg")
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
```

```python
convolution2DBlur = cv2.filter2D(src=imgRGB,ddepth=-1,kernel=kernel)
display(convolution2DBlur)
```

```python
#create new gray scale img
imgBGR = cv2.imread("../DATA/giraffes.jpg",0)
#imgRGB = cv2.cvtColor(imgBGR,cv2.COLOR_BGR2RGB)
display(imgBGR,cmap="gray")
```

```python
HorizontalSobel = cv2.Sobel(src=imgBGR,ddepth=cv2.CV_64F,dx=1,dy=0,ksize=5)
display(HorizontalSobel,cmap="gray")
```

```python
#reload img
imgBGR = cv2.imread("../DATA/giraffes.jpg")
```

```python
#draw histograms for RGB
color = ['b',"g","r"]
for i, col in enumerate(color):
    histr = cv2.calcHist(images=[imgBGR],channels=[i],mask=None,histSize=[256],ranges=[0,256])
    plt.plot(histr,color=col)
plt.title('Histogram')
```
